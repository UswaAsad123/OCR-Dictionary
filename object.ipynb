```python
import cv2
from PIL import Image
import pytesseract
from nltk import ngrams
import requests

def preprocess_image(image_path):
    image = cv2.imread(image_path)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    _, thresholded_image = cv2.threshold(gray_image, 150, 255, cv2.THRESH_BINARY)
    return thresholded_image

def extract_text_from_image(image_path):
    preprocessed_image = preprocess_image(image_path)
    pil_image = Image.fromarray(preprocessed_image)
    extracted_text = pytesseract.image_to_string(pil_image, lang='eng')
    cleaned_text = ''.join([char if ord(char) < 128 else ' ' for char in extracted_text])
    return cleaned_text

def generate_ngrams(text, n):
    words = text.split()
    n_grams = list(ngrams(words, n))
    return n_grams

def get_word_meaning(word):
    # URL for the Free Dictionary API
    url = f'https://api.dictionaryapi.dev/api/v2/entries/en/{word}'
    
    # Make a request to the API
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()
        # Extract meanings from the response
        meanings = []
        for meaning in data[0]['meanings']:
            part_of_speech = meaning['partOfSpeech']
            for definition in meaning['definitions']:
                meanings.append(f"{part_of_speech}: {definition['definition']}")
        return meanings
    else:
        return [f"Word not found or API error: {response.status_code}"]

# Main script
image_path = r'C:\Users\Hamza\Downloads\aiwplain1.jpg'
extracted_text = extract_text_from_image(image_path)
print("Extracted Text:", extracted_text)

# Generate unigrams (n=1)
unigrams = generate_ngrams(extracted_text, 1)
print("Unigrams:", [unigram[0] for unigram in unigrams])

# Prompt the user to input a word from the extracted text
selected_word = input("Enter a word from the extracted text: ")

# Find and print the meaning of the selected word
meanings = get_word_meaning(selected_word)
print(f"Meanings of '{selected_word}':", meanings)


```
